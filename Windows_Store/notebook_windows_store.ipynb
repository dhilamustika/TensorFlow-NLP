```python
import pandas as pd
df = pd.read_csv('windows_store.csv')
df
```

```python
import numpy as np
np.unique(df['Age Rating'])
```

```python
df = pd.DataFrame(df)
ratinglist = df['Age Rating'].tolist()
print("\n For ages 3 and up:", ratinglist.count('For ages 3 and up'),
      "\n For ages 7 and up:", ratinglist.count('For ages 7 and up'),
      "\n For ages 12 and up:", ratinglist.count('For ages 12 and up'),
      "\n For ages 16 and up:", ratinglist.count('For ages 16 and up'),
      "\n For ages 18 and up:", ratinglist.count('For ages 18 and up'),
      "\n Not Rated:", ratinglist.count('Not Rated'))
```

```python
# Remove unnecessary columns
df = df.drop(columns=['Name', 'Price', 'Publisher', 'Date of Release',
                      'Category', 'Size', 'Languages'])
df
```

```python
# One-hot-encoding process
category = pd.get_dummies(df['Age Rating'])
df_cat = pd.concat([df, category], axis=1)
df_cat = df_cat.drop(columns='Age Rating')
df_cat
```

```python
# Preprocessing


# Change to lowercase
sentence = df_cat.Description[1]
sentence_lower = sentence.lower()

# Remove punctuations
import string
sentence_punc = sentence_lower.translate(str.maketrans("","",string.punctuation))
import re
text_punc = re.sub("[^A-Za-z\s']"," ", sentence_punc)  # remove non-alphabet

# Remove stopwords
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(text_punc)
sentence_stopwords = ' '.join([word for word in word_tokens if word not in stop_words])

# Stemming
from nltk.stem import PorterStemmer
porter = PorterStemmer()
word_tokens2 = word_tokenize(sentence_stopwords)
stems = []
for w in word_tokens2:
  stems.append(porter.stem(w))
sentence_stem = ' '.join([word for word in stems])

# Make a function
def text_process(text):
  text = text.lower()
  text = re.sub("[^A-Za-z\s']"," ", text)
  word_tokens = word_tokenize(text)
  text = ' '.join([word for word in word_tokens if word not in stop_words])
  word_tokens2 = word_tokenize(text)
  stemm = []
  for w in word_tokens2:
    stemm.append(porter.stem(w))
    text = ' '.join([word for word in stemm])
  return text

# New dataframe
df_cat['desc_clean'] = df_cat['Description'].apply(lambda x: text_process(x))
df_cat
```

```python
df_new = df_cat.rename(columns={'For ages 3 and up': 'ages_3',
                                'For ages 7 and up': 'ages_7',
                                'For ages 12 and up': 'ages_12',
                                'For ages 16 and up': 'ages_16',
                                'For ages 18 and up': 'ages_18',
                                'Not Rated': 'not_rated'})
```

```python
df_new = df_new[['desc_clean', 'ages_3', 'ages_7', 'ages_12',
                 'ages_16', 'ages_18', 'not_rated']]
df_new
```

```python
# Convert to array
desc_clean = df_new['desc_clean'].values
label = df_new[['ages_3', 'ages_7', 'ages_12', 'ages_16', 'ages_18', 'not_rated']].values
```

```python
# Split the folder into train and test sets
from sklearn.model_selection import train_test_split
desc_clean_train, desc_clean_test, label_train, label_test = train_test_split(desc_clean, label, test_size=0.2)
```

```python
# Use tokenizer 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=5000, oov_token='x')
tokenizer.fit_on_texts(desc_clean_train)
tokenizer.fit_on_texts(desc_clean_test)

seq_train = tokenizer.texts_to_sequences(desc_clean_train)
seq_test = tokenizer.texts_to_sequences(desc_clean_test)

padded_train = pad_sequences(seq_train, padding='post', maxlen=250, truncating='post')
padded_test = pad_sequences(seq_test, padding='post', maxlen=250, truncating='post')
```

```python
# Build the model
import tensorflow as tf
from keras.layers import LSTM

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=16),
#    tf.keras.layers.Bidirectional(LSTM(32, return_sequences=True), input_shape=(32, 250, 6)),
#    tf.keras.layers.LSTM(64, dropout=0.2, activation='relu', return_sequences=True, go_backwards=True),
    tf.keras.layers.LSTM(64, dropout=0.2),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(6, activation='softmax')
])

# Compile the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.8):
      print("\nAccuracy has reached >80%!")
      self.model.stop_training = True

callbacks = myCallback()

history = model.fit(padded_train, label_train, epochs=20, callbacks=[callbacks],
                    validation_data=(padded_test, label_test), verbose=2)
```

```python
# Plot the accuracy of the model
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Plot the loss of the model
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
